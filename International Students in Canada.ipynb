# International Students in Canada

```python
#Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Import Dataset
df = pd.read_csv('Internation_Students_Canada.csv')
```

```python
df.head()
```

```python
#2015
df_2015 = df[['Country of Citizenship','2015']].set_index('Country of Citizenship')
df_2015_Top5 = df_2015.sort_values(by=['2015'], ascending=False).head(5)
df_2015_Top5
```

```python
df_2015_Top5.plot(kind='bar', legend=False, title='Top 5 International Students in Canada 2015')
```

```python
sum_of_2015 = df['2015'].sum()
sum_of_2016 = df['2016'].sum()
sum_of_2017 = df['2017'].sum()
sum_of_2018 = df['2018'].sum()
sum_of_2019 = df['2019'].sum()
sum_of_2020 = df['2020'].sum()
sum_of_2021 = df['2021'].sum()
sum_of_2022 = df['2022'].sum()
sum_of_2023 = df['2023'].sum()
```

```python
year_list = [sum_of_2015, sum_of_2016, sum_of_2017, sum_of_2018, sum_of_2019, sum_of_2020, sum_of_2021, sum_of_2022, 
             sum_of_2023]
```

```python
year_by_year = np.array(year_list)
labels = ['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']

plt.plot(labels, year_by_year,  color = 'b')
plt.title('Total of International Students in Canada 2015-2023')
plt.show()
```
